```python
#import the basic packages
import matplotlib.pyplot as plt
import numpy as np
from numpy import random
```

```python
#Make a little array
x = np.linspace(0,10,10)
y = random.randint(0,100,10)
```

```python
#see what x is
x
```

```python
#see what y is
y
```

```python
plt.plot(x,y, color='g', marker='x', label='random values in green')

plt.xlabel= ('x axis') #not sure why these aren't being displayed but it should work
plt.ylabel= ('y axis')

plt.legend()
plt.tight_layout()

plt.show()

plt.savefig('homework1fig')
```
